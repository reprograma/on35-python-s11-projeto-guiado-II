# 3. Quais foram os valores máximos e mínimos identificados?

```python
import pandas as pd
```

```python
df = pd.read_csv('INMET_MS_ITAQUIRAI_2020.CSV', delimiter=';', skiprows=8, encoding='latin1')
```

```python
colunas = [
    'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)',
    'TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)',
    'TEMPERATURA DO PONTO DE ORVALHO (°C)',
    'UMIDADE RELATIVA DO AR, HORARIA (%)',
    'RADIACAO GLOBAL (Kj/m²)',
    'VENTO, DIREÇÃO HORARIA (gr) (° (gr))',
    'VENTO, VELOCIDADE HORARIA (m/s)'
]

for coluna in colunas:
    df[coluna] = pd.to_numeric(df[coluna], errors='coerce')

for coluna in colunas:
    valor_minimo = df[coluna].min()
    valor_maximo = df[coluna].max()
    print(f'O valor mínimo da coluna "{coluna}" é: {valor_minimo}')
    print(f'O valor máximo da coluna "{coluna}" é: {valor_maximo}')
```

# 5. Qual foi o impacto da normalização de uma coluna específica?
Para a normalização eu considerei a coluna de "UMIDADE RELATIVA DO AR, HORARIA (%)'].fillna(df['UMIDADE RELATIVA DO AR, HORARIA (%)".

Como a normalização serve considerar valores de colunas para um mesmo intervalo, geralmente entre 0 e 1, como um parâmetro, este foi o impacto, proporcionando um "padrão" e sendo mais fácil fazer um comparativo dos dados.

```python
df['UMIDADE RELATIVA DO AR, HORARIA (%)'].fillna(df['UMIDADE RELATIVA DO AR, HORARIA (%)'].mean(), inplace=True)
```
