Organize e limpe os dados, removendo duplicatas e normalizando quando necessário.

2. *Renomeação e Ajuste de Colunas*

Renomeie colunas e ajuste os tipos de dados conforme necessário para garantir a consistência e clareza.

```python
#Renomer as colunas
df.rename(columns={
    'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'Precipitacao_Total_mm',
    'TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)': 'Temperatura_C',
    'UMIDADE RELATIVA DO AR, HORARIA (%)': 'Umidade_Relativa',
    'RADIACAO GLOBAL (Kj/m²)': 'Radiacao_Global_Kj_m2',
    'VENTO, DIREÇÃO HORARIA (gr)': 'Direcao_Vento_graus',
    'VENTO, VELOCIDADE HORARIA (m/s)': 'Velocidade_Vento_m_s'
}, inplace=True)

print("Colunas renomeadas:")
print(df.columns)
```

```python
#Ajustar os tipos de dados da coluna 'Data' - Alterar data para datetime e 'Hora UTC' para time
df['Data'] = pd.to_datetime(df['Data'], format='%Y/%m/%d')
```

```python
df['Data'] = df['Data'].dt.strftime('%d/%m/%Y')
df
```

```python
#Ajustar a coluna 'Hora UTC' para o tipo de dado time
df['Hora UTC'] = pd.to_datetime(df['Hora UTC'], format='%H%M UTC', errors='coerce').dt.strftime('%H:%M')
```

```python
df.head()
```

```python
df['Data e Hora'] = df['Data'] + ' ' + df['Hora UTC']
```

```python
df.head()
```

```python
df['Data e Hora'] = pd.to_datetime(df['Data e Hora'], format='%d/%m/%Y %H:%M', errors='coerce')
```

```python
df.head()
```

```python
df['Data e Hora BR'] = df['Data e Hora'].dt.tz_localize('UTC').dt.tz_convert('America/Sao_Paulo')
```

```python
df.head()
```

```python
df['Data e Hora BR'] = df['Data e Hora BR'].dt.strftime('%d/%m/%Y %H:%M')
```

```python
df.head()
```

```python
#Reduzir o Dataframe para melhor análise dos dados 
df.rename(columns={'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'Precipitação Total'}, inplace=True)
```

```python
#Remover colunas
df.drop('Data e Hora', axis=1, inplace=True)
```

3. *Transformações e Criação de Novas Colunas*

Realize transformações relevantes nos dados, como criar novas colunas derivadas de outras existentes.


```python
#Converter as colunas para formato numérico
df['Temperatura_C)'] = pd.to_numeric(df['Temperatura_C'], errors='coerce')
df['C'] = pd.to_numeric(df['Umidade_Relativa'], errors='coerce')
```

```python
#Criar uma nova coluna 
df['Sensação_Térmica'] = df['Temperatura_C)'] - ((100- df['Umidade_Relativa'])/5)
```

2. *Agrupamento e Sumarização*

Agrupe os dados para identificar padrões e tendências, gerando sumarizações que permitam uma análise mais profunda.

```python
#Realizando o agrupamento da coluna umidade relativa e calculando a média da Sensação térmica'
agrupado = df.groupby('Umidade_Relativa')['Sensação_Térmica'].mean()
print(agrupado)
```

#### **Persistência dos Resultados no SQLite**

1. **Criação do Banco de Dados:**
   - Estabeleça um banco de dados SQLite para armazenar os resultados das análises.

```python
#Importando o sqlite - Para criar o banco de dados
import sqlite3
import pandas as pd
```

```python
#Conectar ao banco de dados (ou criando, se não existir)
#Criar um cursor para interagir com o banco de dados
conn = sqlite3.connect('dados_meteorologicos.db')
cursor = conn.cursor()

```

```python
#Criar a tabela
cursor.execute('''
CREATE TABLE IF NOT EXISTS dados_meteorologicos (
    data TEXT,
    hora_utc TEXT,
    precipitacao_total_mm TEXT,
    pressao_atm_estacao_mb TEXT,
    pressao_atm_max_mb TEXT,
    pressao_atm_min_mb TEXT,
    radiacao_global_kjm2 TEXT,
    temperatura_ar_c TEXT,
    temperatura_orvalho_c TEXT,
    temperatura_max_c TEXT,
    temperatura_min_c TEXT,
    temperatura_orvalho_max_c TEXT,
    temperatura_orvalho_min_c TEXT,
    umidade_rel_max FLOAT,
    umidade_rel_min FLOAT,
    umidade_rel_ar FLOAT,
    vento_direcao_gr FLOAT,
    vento_rajada_max_ms TEXT,
    vento_velocidade_ms TEXT
)
''')
conn.commit()
```
