#### 7. Como os dados foram agrupados e quais insights foram gerados?

Utilizei a função groupby para agregar as variáveis por mês e tirei a média mensal dessas variáveis. Nota-se que, em relação a precipitação, os meses de fevereiro e de junho são os que apresentam a maior quantidade de chuva. Também fevereiro apresentou a maior temperatura média. Em relação a temperatura ponto de orvalho, o mês de janeiro foi destaque. Novembro foi o mês com maior radiação. Em relação a direção em que os ventos vêm, é notório que ela varia consideralmente de um mês para o outro. Além disso, outubro é o mês de maior velocidade dos ventos, seguido de julho e junho, respectivamente. Essas informações são úteis para a área agrícula, para criação de políticas públicas e em diversos outros contextos. 

```python
df['Ano'] = df['Data e Hora'].dt.year
df['Mês'] = df['Data e Hora'].dt.month
```

```python
media_prec = df.groupby('Mês')['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'].mean()
media_prec
```

```python
media_temp_bulbo = df.groupby('Mês')['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'].mean()
media_temp_bulbo
```

```python
media_temp_porvalho = df.groupby('Mês')['TEMPERATURA DO PONTO DE ORVALHO (°C)'].mean()
media_temp_porvalho
```

```python
media_umidade = df.groupby('Mês')['UMIDADE RELATIVA DO AR, HORARIA (%)'].mean()
media_umidade
```

```python
media_radiacao = df.groupby('Mês')['RADIACAO GLOBAL (Kj/m²)'].mean()
media_radiacao
```

```python
media_vento_direcao = df.groupby('Mês')['VENTO, DIREÇÃO HORARIA (gr) (° (gr))'].mean()
media_vento_direcao
```

```python
media_vento_velocidade = df.groupby('Mês')['VENTO, VELOCIDADE HORARIA (m/s)'].mean()
media_vento_velocidade
```
